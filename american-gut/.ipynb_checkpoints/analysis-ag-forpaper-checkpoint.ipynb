```python
import calour as ca
ca.set_log_level('INFO')
%matplotlib notebook
```

```python
dat=ca.read_amplicon('./ag.150.min10.16s.nobloom.withtax.biom','map.txt',normalize=10000,filter_reads=1000)
```

```python
fec=dat.filter_samples('BODY_PRODUCT','UBERON:feces')
sal=dat.filter_samples('BODY_PRODUCT','UBERON:saliva')
skin=dat.filter_samples('BODY_PRODUCT','UBERON:sebum')
```

```python
fecc=fec.filter_min_abundance(100)
```

```python
tt=fecc.sort_abundance()
```

```python
tt.plot(gui='qt5')
```

```python
fecc=cah.cluster_features(fec,50)
```

```python
salc=cah.cluster_features(sal,50)
```

```python
skinc=cah.cluster_features(skin,50)
```

```python
fecc.sample_metadata.columns
```

```python
cah.plot_s(fecc,'BMI',gui='qt5')
```

```python
import calour.analysis
```

# skin

```python
dd=calour.analysis.diff_abundance(skinc,'SEX','male','female')
```

```python
cah.plot_s(dd,'SEX',gui='qt5')
```
