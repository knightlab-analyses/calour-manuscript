```python
import calour as ca
%matplotlib notebook
```

```python
dat=ca.read_amplicon('./ag.150.min10.16s.nobloom.withtax.biom','map.txt',normalize=10000,filter_reads=1000)
```

```python
fec=dat.filter_samples('BODY_PRODUCT','UBERON:feces')
sal=dat.filter_samples('BODY_PRODUCT','UBERON:saliva')
skin=dat.filter_samples('BODY_PRODUCT','UBERON:sebum')
```

```python
fecc=fec.filter_min_abundance(100)
```
