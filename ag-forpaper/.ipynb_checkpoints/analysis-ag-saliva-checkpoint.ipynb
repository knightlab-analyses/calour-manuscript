```python
import calour as ca
%matplotlib notebook
```

```python
dat=ca.read_amplicon('./ag.150.min10.16s.nobloom.withtax.biom','map.txt',normalize=10000,filter_reads=1000)
```

```python
fec=dat.filter_samples('BODY_PRODUCT','UBERON:feces')
sal=dat.filter_samples('BODY_PRODUCT','UBERON:saliva')
skin=dat.filter_samples('BODY_PRODUCT','UBERON:sebum')
```

```python
salc=sal.filter_min_abundance(10)
```

```python
del dat
```

```python
tt=salc.cluster_features(10)
```

```python
zz=tt.collapse_taxonomy()
```

```python
zz=zz.sort_abundance()
```

```python
zzz=zz.filter_ids(zz.feature_metadata.index.values[-30:])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = plt.figure()
freq=[]
tax=[]
allmean = zzz.data.mean(axis=0)
for idx,cbact in enumerate(zzz.feature_metadata.index.values):
    ctax = zzz.feature_metadata.taxonomy[cbact].split(';')
    ctax = ';'.join(ctax[-2:])
    tax.append(ctax)
    freq.append(allmean[idx]/100)
plt.barh(np.arange(len(freq)),freq,align='center')
plt.xscale('log')
plt.yticks(np.arange(len(freq)),tax)
plt.tight_layout()
```

```python
f.savefig('figure-saliva-highest.pdf')
```

# load the skin-body-sites saliva cluster

```python
ss=ca.read_amplicon('../skin-body-sites/skin-saliva.biom','../skin-body-sites/skin-saliva_sample.txt',normalize=10000,filter_reads=1000)
```

```python
ss=ss.filter_samples('bodysite','Arm')
```

```python
combined=ss.join_experiments(salc)
```

```python
combined
```

```python
combined=combined.downsample('orig_exp')
```

```python
combined=combined.sort_abundance()
```

```python
combined.plot_sort('orig_exp',gui='qt5')
```

```python
combined.sample_metadata.orig_exp.unique()
```

```python
f = plt.figure()
useseqs=30
freqag=[]
freqsk=[]
tax=[]
combag = combined.filter_samples('orig_exp','ag.150.min10.16s.nobloom.withtax.biom')
combsk = combined.filter_samples('orig_exp','skin-saliva.biom')
allmeanag = combag.data.mean(axis=0)
allmeansk = combsk.data.mean(axis=0)
for idx,cbact in enumerate(combined.feature_metadata.index.values):
    ctax = combined.feature_metadata.taxonomy[cbact].split(';')
    ctax = ';'.join(ctax[-2:])
    tax.append(ctax)
    freqag.append(allmeanag[idx]/100)
    freqsk.append(allmeansk[idx]/100)
plt.barh(np.arange(useseqs),freqag[-useseqs:],height=0.3,align='center')
plt.barh(np.arange(useseqs)+0.3,freqsk[-useseqs:],height=0.3,align='center')
plt.xscale('log')
plt.yticks(np.arange(useseqs),tax[-useseqs:])
plt.legend(['saliva','skin'])
plt.tight_layout()
```

```python
f.savefig('figure-saliva-skin-combined.pdf')
```

```python
plt.figure()
plt.bar(np.arange(len(m1)),m1,width=1,alpha=0.5)
plt.bar(np.arange(len(m2)),m2,width=1,alpha=0.5)
plt.yscale('log')
plt.xlim([0,100])
plt.ylim([0.001,30])
plt.ylabel('Frequency (%)')
plt.xlabel('bacteria')
```
