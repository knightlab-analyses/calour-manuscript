```python
%matplotlib notebook
import heatsequer as hs
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
readfl=open('readspersample.txt')
```

# load original reads per sample
from:
grep -c '>' deblur-neg/split > reads.txt

```python
reads={}
for cline in readfl:
    cline=cline.split(':')
    csamp=cline[0].split('/')[-1]
    creads=int(cline[1])
    reads[csamp]=creads
```

# load deblurred reads per sample
from biom summarize and manually removing header (not needed)

```python
sumfl=open('debneg.sum.forparse.txt')
rat={}
dreads={}
for cline in sumfl:
    try:
        cline=cline.split(':')
        cline[1]=cline[1].strip()
        creads=float(cline[1])
        csamp=cline[0]+'.fasta'
        dreads[csamp]=creads
        if csamp not in reads:
            print('sample %s not in reads' % csamp)
            continue
        orignum=reads[csamp]
        if orignum<500:
            continue
        if 'BLANK' in csamp:
            continue
        rat[csamp]=creads/orignum
    except:
        print('error %s' % cline)
```

```python
plt.figure()
vals=[]
for cval in rat.values():
    vals.append(cval)
vals=np.array(vals)
plt.hist(vals,50)
plt.xlabel('ratio deblur/orig(after trimming) num reads')
plt.ylabel('number of samples')
plt.title('effect of deblurring on number of reads - negative filtering')
plt.savefig('numreads-ratio-neg-filter.pdf')
print(np.sum(vals<0.35))
print(len(vals))
```

```python
for k,v in rat.items():
    if v<0.2:
        print(k)
        print(reads[k])
        print(dreads[k])
```
